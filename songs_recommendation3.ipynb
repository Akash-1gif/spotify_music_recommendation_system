#### Importing ML models:

### Step 6: Songs recommendation using Nearest Neighbours model (From sklearn library):

```python
df_test_1=df_test.drop(['track_id','track_name','track_genre','Cluster_Labels'],axis=1)
df_test_1.head()
```

### Step 6.1: Songs recommendation based on a single song:

##### selecting a song. Based on this song's attribute values, the new songs are recommended:

```python
single_point=df_test_1.iloc[0]
single_point
```

#### Step 6.1a: Predicting the cluster for our favourite song (single_point):

```python
predicted_class=rfc.predict([single_point])[0]
predicted_class
```

a correct classification has been made using Random Forest Classifier (check cell number 6 for clarification). This gives a proper clarification that our model is working appropriately.

#### step 6.1b: Shortlisting songs from the predicted cluster:

```python
df_class=df_train[df_train['Cluster_Labels']==predicted_class]
df_class.head()
```

```python
df_class.shape
```

```python
df_class_1=df_class.drop(['track_id','track_name','track_genre','Cluster_Labels'],axis=1)
df_class_1.head()
```

#### step 6.1c: Adding single_point to the end of the df_class_1, so that we can use it to point out the nearest datapoints to it

```python
df_class_1.loc[99999999]=single_point
df_class_1.tail(5)
```

#### Fetching the row of the single_point:

```python
row_to_query=df_class_1.loc[99999999][['popularity','danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo','track_genre_num']].values.reshape(1, -1)
row_to_query
```

#### step 6.1d: Importing NearestNeighbors from sklearn and creating it's instance:

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neighbors_model=NearestNeighbors(n_neighbors=10)
```

```python
neighbors_model.fit(df_class_1.values)
```

#### step 6.1e: Fetching out the 10 most nearest data points to single_point data point and recommending them as new songs for the user:

```python
distances,indices=neighbors_model.kneighbors(row_to_query)
```

```python
nearest_neighbors=df_class_1.iloc[indices[0]]
```

```python
nearest_neighbors
```

```python
rat=nearest_neighbors.index
```

#### Step 6.1f: Displaying the metadata of the recommended songs:

```python
for i in rat[1:]:
    print(df_class.loc[i])
    print('_______________________________________')
```

```python
distances[0][1:]
```
